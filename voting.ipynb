```python
c1 = input("enter the name of country ")
c2 = input("enter the name of country ")
c3 = input("enter the name of country ")
c4 = input("enter the name of country ")
c5 = input("enter the name of country ")

c1_votes = 0
c2_votes = 0
c3_votes = 0
c4_votes = 0
c5_votes = 0

passport = [6, 7, 8, 9, 10]
num_of_voters = len(passport)
```

```python
while True:
    if passport == []:
        print("voting session over")
        if (
            c1_votes > c2_votes
            and c1_votes > c3_votes
            and c1_votes > c4_votes
            and c1_votes > c5_votes
        ):
            percent = (c1_votes / num_of_voters) * 100
            print(
                c1,
                "has won",
                "with",
                percent,
                "% votes"
            )
            break
        elif (
            c2_votes > c1_votes
            and c2_votes > c3_votes
            and c2_votes > c4_votes
            and c2_votes > c5_votes
        ):
            percent = (c2_votes / num_of_voters) * 100
            print(
                c2,
                "has won",
                "with",
                percent,
                "% votes"
            )
            break
        elif (
            c3_votes > c1_votes
            and c3_votes > c2_votes
            and c3_votes > c4_votes
            and c3_votes > c5_votes
        ):
            percent = (c3_votes / num_of_voters) * 100
            print(
                c3,
                "has won",
                "with",
                percent,
                "% votes",
                "Number of Vote is :",
                c3_votes,
            )
            break
            
        elif (
            c4_votes > c1_votes
            and c4_votes > c2_votes
            and c4_votes > c3_votes
            and c4_votes > c5_votes
        ):
            percent = (c4_votes / num_of_voters) * 100
            print(
                c4,
                "has won",
                "with",
                percent,
                "% votes",
                "Number of Vote is :",
                c4_votes,
            )
            break

        elif (
            c5_votes > c1_votes
            and c5_votes > c2_votes
            and c5_votes > c3_votes
            and c5_votes > c4_votes
        ):
            percent = (c5_votes / num_of_voters) * 100
            print(
                c5,
                "has won",
                "with",
                percent,
                "% votes",
                "Number of Vote is :",
                c5_votes,
            )
            break
    else:
        voter = int(input("enter your passport number = "))
        if voter in passport:
            print("---------------------- You're a register voter -----------------")
            passport.remove(voter)
            vote =input(f"enter your vote {c1} or {c2} or {c3} or {c4} or {c5} =" )
            if vote==c1:
                c1_votes+=1
                print("---------------Thanks for Vote Casting for--",c1)
            elif vote ==c2:
                c2_votes+=1
                print("---------------Thanks for Vote Casting for --",c2)
            elif vote ==c3:
                c3_votes+=1
                print("---------------Thanks for Vote Casting for --",c3)
            elif vote ==c4:
                c4_votes+=1
                print("---------------Thanks for Vote Casting for--",c4)
            elif vote ==c5:
                c5_votes+=1
                print("---------------Thanks for Vote Casting for --",c5)
        else:
            print(">>>>>>>>>>>>>>>You're not listed voter or you've already voted<<<<<<<<<<<<<<<<<<")
```
